# Scenario B: Warehouse Pick-and-Pack

This notebook presents the scenario as an operations case study: analytical bottleneck diagnosis, simulation validation, redesign comparison, stress testing, sensitivity analysis, and cost recommendation.


## Modeling Design

- Six-step serial process with finite-capacity resources.
- Poisson arrivals with exponential inter-arrival times.
- Service-time distributions vary by station: exponential, Erlang-2, or lognormal.
- First 150 items are discarded as warmup to reduce transient bias.
- Each configuration uses 10 replications with reproducible seeds.
- All reported confidence intervals are 95% Student's t intervals.


```python
import pandas as pd
from IPython.display import Image, Markdown, display
from run_analysis import analyse_process, run_all, scenario_artifacts

artifacts = run_all(selected='warehouse')
scenario = artifacts['warehouse']
spec = scenario_artifacts('warehouse')
comparison, cost, stress = artifacts['summary_tables']
best = scenario['best_label']
```

## Analytical Bottleneck Check

The first pass uses station utilisation and Erlang-C queueing estimates before simulation. That makes the bottleneck diagnosis explainable rather than purely empirical.


```python
analysis = analyse_process(spec['configs']['Current State']).copy()
analysis['Utilisation_rho'] = analysis['Utilisation_rho'].round(3)
analysis['Avg_Wait_Time_min'] = analysis['Avg_Wait_Time_min'].replace(float('inf'), pd.NA)
display(analysis[['Step', 'Lambda_in', 'Mu_per_server', 'Servers', 'Utilisation_rho', 'Avg_Wait_Time_min', 'Is_Bottleneck']])
```

## Simulation Results

The table below compares the current state and redesigns using replicated simulation outputs. Cycle-time and throughput estimates include 95% confidence intervals.


```python
cols = ['config', 'bottleneck_rho', 'cycle_time_min_mean', 'cycle_time_min_ci_low', 'cycle_time_min_ci_high', 'throughput_hr_mean', 'bottleneck_queue_mean']
display(comparison[cols].round(2))
display(Markdown(f'**Recommended redesign:** {best}'))
```

## Cost And Stress Testing

Redesigns are evaluated as business decisions, not just queueing improvements. The cost table converts throughput gains into net benefit, while the stress test checks behavior at +50% demand.


```python
display(cost.round(2))
display(stress.round(2))
```
